```python
import pandas as pd
import json
```

```python
df = pd.read_csv(
    r"/Data/processed_healthcare_dataset.csv",
)

print(df.shape)

df.head()
```

```python
def create_qql_entry(row):
    
    retrieval_text = str(row["retrieval_text"]).replace('"', '\\"')
    
    metadata = {
        "patient_query": str(row["patient_query"]),
        "doctor_response": str(row["doctor_response"]),
        "query_tokens": int(row["query_tokens"]),
        "response_tokens": int(row["response_tokens"])
    }

    metadata_json = json.dumps(metadata, indent=2)

    qql_entry = f"""
INSERT INTO healthcare_conversations
TEXT "{retrieval_text}"

METADATA {metadata_json}
"""

    return qql_entry
```

```python
qql_entries = []

for _, row in df.iterrows():
    
    entry = create_qql_entry(row)
    
    qql_entries.append(entry)
```

```python
print(qql_entries[0])
```

```python
full_qql_script = "\n".join(qql_entries)
```

```python
save_querys = r"C:\Users\arsha\OneDrive\Desktop\Healthcare Retrieval using qql\Data\healthcare_conversations.qql"

with open(save_querys, "w", encoding="utf-8") as file:
    
    file.write(full_qql_script)

print("QQL file generated successfully!")
```
