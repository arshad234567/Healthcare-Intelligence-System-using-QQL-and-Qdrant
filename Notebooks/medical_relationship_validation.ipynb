```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
```

```python
#load filtered relationships

filtered_relationships=pd.read_csv(
    r"C:\Users\arsha\OneDrive\Desktop\Healthcare Retrieval using qql\Data\graph_ready_relationships.csv"
)

print(filtered_relationships.shape)

display(filtered_relationships.head())
```

```python
valid_medical_relations={

    "dizziness":[
        "hypoglycemia",
        "blood pressure",
        "bppv",
        "benign paroxysmal positional vertigo",
        "peripheral vertigo",
        "ischemia",
        "heart rhythm",
        "anemia",
        "migraine"
    ],

    "cough":[
        "asthma",
        "bronchitis",
        "sinusitis",
        "infection",
        "cold",
        "flu",
        "covid",
        "pneumonia",
        "allergy"
    ],

    "fever":[
        "infection",
        "flu",
        "viral infection",
        "typhoid",
        "covid",
        "malaria",
        "dengue"
    ],

    "headache":[
        "migraine",
        "blood pressure",
        "stress",
        "sinusitis",
        "dehydration"
    ],

    "nausea":[
        "pregnancy",
        "food poisoning",
        "infection",
        "gastritis",
        "migraine"
    ],

    "chest pain":[
        "heart attack",
        "cardiac disease",
        "angina",
        "acid reflux"
    ]
}
```

```python
#store validated relationships

validated_relationships=[]
```

```python
#medical semantic filtering

for _,row in filtered_relationships.iterrows():

    source=row["source"]

    target=row["target"]

    relationship=row["relationship"]

    count=row["count"]

    if source in valid_medical_relations:

        if target in valid_medical_relations[source]:

            validated_relationships.append({

                "source":source,
                "target":target,
                "relationship":relationship,
                "count":count

            })
```

```python
#create validated dataframe

validated_relationships_df=pd.DataFrame(
    validated_relationships
)

print(validated_relationships_df.shape)

display(
    validated_relationships_df.head(10)
)
```

```python
#check dizziness relationships

dizziness_relations=validated_relationships_df[
    validated_relationships_df["source"]=="dizziness"
]

display(dizziness_relations)
```

```python
cough_relations=validated_relationships_df[
    validated_relationships_df["source"]=="cough"
    ]
display(cough_relations)
```

```python
fever_relations=validated_relationships_df[
    validated_relationships_df["source"]=="fever"
    ]
display(fever_relations)
```

```python
#top validated relationships

validated_relationships_df=validated_relationships_df.sort_values(
    by="count",
    ascending=False
)

display(
    validated_relationships_df.head(30)
)
```
